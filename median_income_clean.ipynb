```python
import pandas as pd

income = pd.read_csv("acs5_b19013_medianhhincomebycounty.csv")

income.columns = income.columns.str.strip()

income['CountyName'] = (
    income['CountyName']
    .str.replace(" County", "", regex=False)
    .str.replace(" city", "", regex=False)
    .str.strip()
)

income = income[['Year', 'CountyName', 'TotalEstimate']]
income = income.rename(columns={
    'TotalEstimate': 'MedianIncome'
})


income_2019 = income[income['Year'] == 2019].reset_index(drop=True)
income_2020 = income[income['Year'] == 2020].reset_index(drop=True)


income_2019 = income_2019.sort_values('CountyName')
income_2020 = income_2020.sort_values('CountyName')


income_2019.to_csv("va_income_2019_clean.csv", index=False)
income_2020.to_csv("va_income_2020_clean.csv", index=False)


print("2019 Income Data:")
print(income_2019.head(), "\n")

print("2020 Income Data:")
print(income_2020.head())
```
