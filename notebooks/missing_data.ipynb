```python
import pandas as pd

df19 = pd.read_csv("broadband_data_2019_clean.csv")
df20 = pd.read_csv("broadband_data_2020_clean.csv")

missing_in_2020 = set(df19["COUNTY NAME"]) - set(df20["COUNTY NAME"])
missing_in_2019 = set(df20["COUNTY NAME"]) - set(df19["COUNTY NAME"])

print("Missing in 2020:")
print(missing_in_2020)

print("\nMissing in 2019:")
print(missing_in_2019)
```
