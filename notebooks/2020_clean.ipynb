```python
import pandas as pd
import numpy as np

df = pd.read_csv("broadband_data_2020October.csv")

df.columns = df.columns.str.strip()
df['ST'] = df['ST'].str.strip()

df = df[df['ST'] == 'VA']

df = df.replace("-", np.nan)
df = df.replace(" - ", np.nan)
df = df.replace(" -   ", np.nan)

df = df.dropna()

df = df.reset_index(drop=True)

print(df.head(150))

df.to_csv("broadband_data_2020_clean.csv", index=False)
```
