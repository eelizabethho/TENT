```python
import pandas as pd

df = pd.read_csv("broadband_data_2019November.csv")

df = df.dropna()
df.columns = df.columns.str.strip()
df['ST'] = df['ST'].str.strip()
df = df[df['ST'] == 'VA']
df = df.reset_index(drop=True)

print(df.head(150))

df.to_csv("broadband_data_2019_clean.csv", index=False)
```
