```python
import pandas as pd
import numpy as np

df = pd.read_csv('broadband_data_opendatachallenge.csv', encoding='latin1')

df.replace(['NULL', 'null', '', ' '], np.nan, inplace=True)

df = df[df['State'] == 'Virginia']
df.dropna(axis=0, how='any', inplace=True)

df.reset_index(drop=True, inplace=True)

print(df.head(100))
df.to_csv('broadband_data_opendatachallenge.csv', index=False)

```
